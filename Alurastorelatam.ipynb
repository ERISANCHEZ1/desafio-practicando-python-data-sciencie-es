### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# ingresos totales de cada tienda
ingresos = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

df_ingresos = pd.DataFrame(list(ingresos.items()), columns=["Tienda", "Ingreso Total"])
print(df_ingresos)
```

# 2. Ventas por categoría

```python
ventas_categoria_t1 = tienda['Categoría del Producto'].value_counts()
ventas_categoria_t2 = tienda2['Categoría del Producto'].value_counts()
ventas_categoria_t3 = tienda3['Categoría del Producto'].value_counts()
ventas_categoria_t4 = tienda4['Categoría del Producto'].value_counts()

print(" Categoría más popular por tienda:")
print("Tienda 1:", ventas_categoria_t1.idxmax(), "-", ventas_categoria_t1.max(), "ventas")
print("Tienda 2:", ventas_categoria_t2.idxmax(), "-", ventas_categoria_t2.max(), "ventas")
print("Tienda 3:", ventas_categoria_t3.idxmax(), "-", ventas_categoria_t3.max(), "ventas")
print("Tienda 4:", ventas_categoria_t4.idxmax(), "-", ventas_categoria_t4.max(), "ventas")
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt

categorias_populares = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Categoría más popular": [
        ventas_categoria_t1.idxmax(),
        ventas_categoria_t2.idxmax(),
        ventas_categoria_t3.idxmax(),
        ventas_categoria_t4.idxmax()
    ],
    "Ventas": [
        ventas_categoria_t1.max(),
        ventas_categoria_t2.max(),
        ventas_categoria_t3.max(),
        ventas_categoria_t4.max()
    ]
})
etiquetas = [
    f"{cat} ({tienda})"
    for cat, tienda in zip(categorias_populares["Categoría más popular"], categorias_populares["Tienda"])
]

plt.figure(figsize=(8, 6))
plt.pie(categorias_populares["Ventas"],
        labels=etiquetas,
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.Paired.colors)
plt.title("Distribución de las Categorías más Populares por Tienda")
plt.axis('equal')
plt.show()
```

# 3. Calificación promedio de la tienda


```python
#3 Calificación promedio de cada tienda
promedio_califificacion_tienda = {
    "Tienda 1:": tienda["Calificación"].mean(),
    "Tienda 2:": tienda2["Calificación"].mean(),
    "Tienda 3:": tienda3["Calificación"].mean(),
    "Tienda 4:": tienda4["Calificación"].mean(),

}
df_calificaciones = pd.DataFrame.from_dict(promedio_califificacion_tienda, orient='index', columns=["Calificación Promedio"])
df_calificaciones.reset_index(inplace=True)
df_calificaciones.rename(columns={"index": "Tienda"}, inplace=True)

print(df_calificaciones)
```

```python
import matplotlib.pyplot as plt

# Diccionario con calificaciones promedio por tienda
promedios_calificacion = {
    "Tienda 1": tienda["Calificación"].mean(),
    "Tienda 2": tienda2["Calificación"].mean(),
    "Tienda 3": tienda3["Calificación"].mean(),
    "Tienda 4": tienda4["Calificación"].mean()
}

plt.figure(figsize=(8, 5))
bars = plt.bar(promedios_calificacion.keys(), promedios_calificacion.values(), color='skyblue')


for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.1, f"{height:.2f}", ha='center', fontsize=10, color='black')


plt.title("Calificación Promedio por Tienda")
plt.ylabel("Calificación Promedio")
plt.ylim(0, 5.5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

# 4. Productos más y menos vendidos

```python
# Agrupar y contar por producto
ventas_t1 = tienda["Producto"].value_counts()
ventas_t2 = tienda2["Producto"].value_counts()
ventas_t3 = tienda3["Producto"].value_counts()
ventas_t4 = tienda4["Producto"].value_counts()

print("Productos más y menos vendidos:")

print("\nTienda 1:")
print("Más vendido:", ventas_t1.idxmax(), "-", ventas_t1.max(), "ventas")
print("Menos vendido:", ventas_t1.idxmin(), "-", ventas_t1.min(), "ventas")

print("\nTienda 2:")
print("Más vendido:", ventas_t2.idxmax(), "-", ventas_t2.max(), "ventas")
print("Menos vendido:", ventas_t2.idxmin(), "-", ventas_t2.min(), "ventas")

print("\nTienda 3:")
print("Más vendido:", ventas_t3.idxmax(), "-", ventas_t3.max(), "ventas")
print("Menos vendido:", ventas_t3.idxmin(), "-", ventas_t3.min(), "ventas")

print("\nTienda 4:")
print("Más vendido:", ventas_t4.idxmax(), "-", ventas_t4.max(), "ventas")
print("Menos vendido:", ventas_t4.idxmin(), "-", ventas_t4.min(), "ventas")
```

```python
import matplotlib.pyplot as plt

def graficar_top5(ventas, titulo):
    top5 = ventas.head(5)
    plt.figure(figsize=(8, 4))
    top5.sort_values().plot(kind='barh', color='lightgreen')
    plt.title(f"Top 5 productos más vendidos - {titulo}")
    plt.xlabel("Cantidad de ventas")
    plt.tight_layout()
    plt.show()

# Graficar para cada tienda
graficar_top5(ventas_t1, "Tienda 1")
graficar_top5(ventas_t2, "Tienda 2")
graficar_top5(ventas_t3, "Tienda 3")
graficar_top5(ventas_t4, "Tienda 4")
```

```python
def graficar_top5(ventas, titulo):
    bottom5 = ventas.sort_values().head(5)  # ordenar de menor a mayor y tomar los 5 primeros
    plt.figure(figsize=(8, 4))
    bottom5.plot(kind='barh', color='blue')
    plt.title(f"Productos menos vendidos - {titulo}")
    plt.xlabel("Cantidad de ventas")
    plt.tight_layout()
    plt.show()

graficar_top5(ventas_t1, "Tienda 1")
graficar_top5(ventas_t2, "Tienda 2")
graficar_top5(ventas_t3, "Tienda 3")
graficar_top5(ventas_t4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
envio_promedio = {
    "Tienda 1": tienda["Costo de envío"].mean(),
    "Tienda 2": tienda2["Costo de envío"].mean(),
    "Tienda 3": tienda3["Costo de envío"].mean(),
    "Tienda 4": tienda4["Costo de envío"].mean(),
}

print(envio_promedio)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_envio = pd.DataFrame.from_dict(envio_promedio, orient='index', columns=['Costo Promedio'])
df_envio.index.name = 'Tienda'
df_envio.reset_index(inplace=True)

plt.figure(figsize=(8, 5))
plt.fill_between(df_envio['Tienda'], df_envio['Costo Promedio'], color='lightgreen', alpha=0.6)
plt.plot(df_envio['Tienda'], df_envio['Costo Promedio'], marker='o', color='green')
plt.title('Costo de Envío Promedio por Tienda')
plt.ylabel('Costo promedio ($)')
plt.xlabel('Tienda')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(
    data=df_ventas, x='lon', y='lat', fill=True, cmap="YlOrRd", thresh=0.05, levels=100
)
plt.title('Mapa de Calor de la Concentración de Ventas')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.tight_layout()
plt.show()
```

Analizar comparativamente el desempeño de las cuatro tiendas considerando cinco aspectos estratégicos:

  •	Facturación total

  •	Categorías de productos más populares

  •	Calificación promedio de clientes

  •	Productos más y menos vendidos

## ***Análisis general***

# **1. Facturación total:**

  •	Tienda 1 presenta la mayor facturación con más de 1,15 mil millones.
  
  •	Tienda 4 es la que menos factura con aproximadamente 1.03 mil millones, siendo la única por debajo del promedio.

# **2. Categorías más populares**

  •	Tienda 4, aunque tiene algunos productos destacados, presenta varios productos de bajo rendimiento con bajo volumen de ventas, lo cual puede reflejar menor rotación de inventario.

  •	Según los datos relacionados y obtenido en el proceso la Tienda 4 presenta el costo promedio de envío más alto, lo que puede afectar la competitividad del canal.
